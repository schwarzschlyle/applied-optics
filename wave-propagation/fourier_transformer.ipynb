```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.color import rgb2hsv, rgb2gray, rgb2yuv
from skimage import color, exposure, transform
from skimage.exposure import equalize_hist
from math import *
from PIL import Image
```

```python
dark_image = imread('hexagon.jpg')
```

```python
dark_image_grey = rgb2gray(dark_image)
plt.figure(num=None, figsize=(8, 6), dpi=80)
plt.imshow(dark_image_grey, cmap='gray');
```

```python
dark_image_grey_fourier = np.fft.fftshift(np.fft.fft2(dark_image_grey))
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80)
plt.imshow((abs(dark_image_grey_fourier)), cmap='gray');
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80)
plt.imshow((np.angle(dark_image_grey_fourier)), cmap='gray');
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80)
test_im = (abs(dark_image_grey_fourier)*(np.exp(1j*np.angle(dark_image_grey_fourier))))
plt.imshow(abs(np.fft.ifft2(test_im)), cmap='gray')
```
